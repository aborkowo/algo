## Git i Github - zdalne repozytorium kodu


Uporządkowana praca nad kodem źródłowym wymaga kontroli i śledzenia wprowadzonych zmian.
Dodatkowo w projektach, w których pracuje jednocześnie wiele osób, konieczny jest system umożliwiający dzielenie wspólnej bazy kodu w całym zespole. 
Niezbędne staje się też rozstrzyganie konfliktów w zmianach wprowadzonych przez programistów. 

W tym celu zostały stworzone systemy kontroli wersji.
Obecnie najpopularniejszym i dodatkowo darmowym systemem kontroli wersji jest git:

https://git-scm.com/

W internecie pojawiło się kilku usługodawców oferujących zdalne utrzymywanie repozytoriów kodu i zarządzanie projektami.
Najpopularniejsze: github, gitlab, gerrit. 
Wiele firm, instytucji, wolnych strzelców (freelancer) oraz hobbistów korzysta z takich rozwiązań.
W celu zachowania kopii bezpieczeństwa i możliwości dzielenia się kodem z innymi programistami lub przyszłymi pracodawcami pracodawcami proponuję założyć konto na Git Hub.


https://help.github.com/en/github/getting-started-with-github

#### Klucze SSH - przydatne informacje

- Generaja kluczy

    ssh-keygen -t rsa -b 4096 -C "<Email@address>"


- Ustawienie dostępu do githuba przez ssh (zamiast URL)

    git remote set-url origin git@github.com:<USERNAME>/<Reopsitory>.git


- Weryfikaja działania SSH

    ssh -T git@github.com


- Dodanie klucza do ssh agenta - eliminuje konieczność podawania hasła do klucza ssh przy pracy ze zdalnym repozytorium

    eval `ssh-agent -s`     
    ssh-add ~/.ssh/id_rsa

# Podstawowe operacje Git

Pobranie zdalnego repozytorium

    git clone git@Server:Nazwa_repozytorium

Serwer np.: github.com

Aktualizacja lokalnego repozytorium o zmiany na zdalnym repozytorium

    git pull

Aktualizacja zdalnego repozytorium o zmiany lokalne

    git push

Sprawdzenie, czy są zmodyfikowane pliki

    git status

Dodanie zmodyfikowanych lub nowych plików

    git add -u
    git add <Nazwa pliku>

Zatwierdzenie zmian w repozytorium

    git commit -m "Komentarz do zmiany"
    
Podgląd zmian w repozytorium

    git log -<N ostatnich zmian>
    
Porównanie zmian w repozytorium

    git diff 
    git diff HEAD^ HEAD
    gid diff <#commit 1> <#commit 2>
