# Algorytm quick sort

Algorytm sortujący bazujący na technice "dziel i zwyciężaj"
1. Podział na dwie podtablice 
 - Wybieramay jeden element tablicy, 
  - elementy mniejsze od wybranego umieszczamy po lewej stronie
  - elementy większe umieszczamy po prawej stronie

2. Wywoałenie rekurencyjne 
 - sortujemy części tablicy z lewej i prawej strony elementu
 - podtablice mogą mieć różne rozmiary, 
    w szczególności jedna z tablic może być pusta, a druga może zawierać wszystkie pozostałe elementy
3. Złożenie wyników
 - po wywołaniu rekurencyjnym tablica jest już posortowana
4. Przypadek bazowy
 - tablica z jednym elementem

## Porównanie z merge sort
- nie potrzebuje dodatkowej pamięci na kopie tablicy
- czas działania zależy od ułożenia danych wejściowych i wyboru elementu startowego
 - podproblemy mogą być znacząco różnych rozmiarów 
 - w najgorszym przypadku $O(N_{2})$
 - w średnim $O(Nlog_{2}(N))$
